### Cleaning
Dejamos los datos crudos por un rato y utilizamos otra variable para trabajar

```python
dsCredit01 = dsCreditRaw
```

Eliminar columna ID

```python
dsCredit01 = dsCredit01.drop("ID", axis=1)
dsCredit01.shape
```

Cambio de tipo variable SEX de numérica continua a tipo categórica

```python
dsCredit01["SEX"] = dsCredit01["SEX"].astype(str)
dsCredit01.SEX.replace(['1','2'],
                       ['M','F'], 
                       inplace=True)
dsCredit01["SEX"] = dsCredit01["SEX"].astype("category")
print(dsCredit01.SEX.unique())
```

Cambio de tipo variable EDUCATION de numérica continua a tipo categórica

```python
dsCredit01["EDUCATION"] = dsCredit01["EDUCATION"].astype(str)
dsCredit01.EDUCATION.replace(['1','2','3','4','5','6','0'],
                             ['Graduate School','University','High School','Others','Others','Others','Others'], 
                             inplace=True)
dsCredit01["EDUCATION"] = dsCredit01["EDUCATION"].astype("category")
print(dsCredit01.EDUCATION.unique())
```

Cambio de tipo variable MARRIAGE de numérica continua a tipo categórica

```python
dsCredit01["MARRIAGE"] = dsCredit01["MARRIAGE"].astype(str)
dsCredit01.MARRIAGE.replace(['1','2','3','0'],
                            ['married','single','divorce','other'],
                            inplace=True)
dsCredit01["MARRIAGE"] = dsCredit01["MARRIAGE"].astype("category")
print(dsCredit01.MARRIAGE.unique())
```

Cambio de tipo variable "default payment next month" de numérica continua a tipo categórica y además cambio de nombre a "DEFAULT_PAYMENT"

```python
dsCredit01.rename(columns={"default payment next month":"DEFAULT_PAYMENT"}, inplace=True)
dsCredit01["DEFAULT_PAYMENT"] = dsCredit01["DEFAULT_PAYMENT"].astype(str)
dsCredit01["DEFAULT_PAYMENT"] = dsCredit01["DEFAULT_PAYMENT"].astype("category")
print(dsCredit01.DEFAULT_PAYMENT.unique())
```
